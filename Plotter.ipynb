```python
def open_json(filename):
    with open(filename) as json_file:
        filedrone = json.load(json_file)

    xsim = []
    ysim = []
    zsim = []

    length = len(filedrone["trajectory"])
    
    timestamp=filedrone["timeline_sec"][1]-filedrone["timeline_sec"][0]
    
    time=filedrone["timeline_sec"]

    j = 0
    for j in range(3):
        k = 0
        if j == 0:
            for k in range(length):
                xr = filedrone["trajectory"][k][j]
                xsim.append(xr)

        if j == 1:
            for k in range(length):
                yr = filedrone["trajectory"][k][j]
                ysim.append(yr)

        if j == 2:
            for k in range(length):
                zr = filedrone["trajectory"][k][j]
                zsim.append(zr)
    return xsim,ysim,zsim,length,timestamp,time

    
```

```python
def opendata_fromclover(filename):
    x=[]
    y=[]
    z=[]
    t=[]
    f=open(filename)
    stroka=' '
    xx=''
    yy=''
    zz=''
    tt=''
    count=0
    while (stroka!=''):
        stroka=f.readline()
    #     print(stroka)
        xx=''
        yy=''
        zz=''
        tt=''
        i=0
        for i in range(len(stroka)-1):
        
            if stroka[i]==' ':
                count+=1
            if count==0:
                xx=xx+stroka[i]
                if (stroka[i]==' '):
                    break
            if count==1:
                yy=yy+stroka[i+1]
            if count==2:
                zz=zz+stroka[i+1]
            if count==3:
                tt=tt+stroka[i+1]
        count=0
        if (xx=='')or (yy=='') or (zz=='') or (tt==''):
            break
        xxx=float(xx)
        yyy=float(yy)
        zzz=float(zz)
        ttt=float(tt)
        x.append(xxx)
        y.append(yyy)
        z.append(zzz)
        t.append(ttt)

    return x,y,z,t
    
```
